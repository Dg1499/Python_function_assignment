A function is a collection of lines of code that accomplishes a certain task.

**Q. Explain the concept of function arguments and parameters in Python.**

```python
#A parameter is the variable listed inside the parentheses in the function definition.
#An argument is the value that are sent to the function when it is called.
def add(num1,num2):       #num1 and num2 are parameters
  return num1+num2
print(add(2.3,12.34))     #2.3 and 12.34 are arguments
```

```python
add(2.3,24.45,12)
```

```python
#for above problem here we can introduce two concepts
#*args (Non Keyword Arguments) --> allow us to pass the variable number of non keyword arguments to function.
#The arguments are passed as a tuple.
#**kwargs (Keyword Arguments) --> allows us to pass the variable length of keyword arguments to the function.
#The arguments are passed as a dictionary.
#*args (Non Keyword Arguments) example
def add1(*args):
  sum=0
  for i in args:
    sum+=i
  return sum
print(add1(2.3,12.34,12,23.45))
```

```python
#*kargs (Non Keyword Arguments) example
def add2(**kargs):
  sum=0
  for i in kargs.values():
    sum+=i
  return sum
print(add2(a=2.3,b=12.34,c=12,d=23.45))
```

```python
#Python’s `reduce()` implements a mathematical technique commonly known as `folding` or `reduction`.
#It is a process where we can reduce a list of items to a single value.
from functools import reduce
numbers = [1, 2, 3, 4, 5]
total_sum = reduce(lambda x, y: x + y, numbers)
print(total_sum)
```

**Q.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.**

```python
def sumofEvennumbers():
  n=int(input("Enter the number of elements of list:"))
  list1=[int(input(f"Enter the {i+1}th element: ")) for i in range(n)]
  list2=[i for i in list1 if i%2==0]
  sum=0
  for i in list2:
    sum+=i
  return sum
```

```python
print("Sum of even numbers from list:",sumofEvennumbers())
```

**Q.Write a Python function that checks if a given number is prime or not from 1 to 200**

```python
import math
def primeChecker(num):
  if num==1:
    return False
  for i in range(2,(int(math.sqrt(num)))):
    if num%i==0:
      return False
  return True
list5=[i for i in range(1,200) if primeChecker(i)]
print(list5)
```

**Q. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.**

```python
def converter(c):
  f=(c*(9/5))+32
  return round(f,2)
celsiuslist = [-44,-40,-23,-1,10,21,33]
farenhitelist = list(map(converter, celsiuslist))
print(farenhitelist)

```

```python
celsiuslist = [-44,-40,-23,-1,10,21,33]
farenhitelist = list(map(lambda c:round(((c*(9/5))+32),2), celsiuslist))
print(farenhitelist)
```
